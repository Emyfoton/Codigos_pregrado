```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Datos celda solar

#Parte 1
d= np.array([30, 40, 50, 60, 70, 80, 90, 100])
vp1= np.array([2.2, 1.7, 1.1, 0.8, 0.7, 0.05, 0.04, 0.03])
cp1= np.array([0.17, 0.12, 0.09, 0.06, 0.05, 0.03, 0.02, 0.02])

#Parte 2
vp2= np.array([0.3, 1.03, 1.4, 1.82, 1.97, 2.02, 2.04, 2.05, 2.06])
ip2= np.array([0.08, 0.08, 0.08, 0.07, 0.5, 0.03, 0.02, 0.01, 0])
```

```python

plt.scatter(d, vp1, c="darkslateblue")
plt.title("Voltaje vs Distancia Celda solar")
plt.xlabel("Distancia (cm)")
plt.ylabel("Voltaje V")
plt.show()
```

```python
plt.scatter(d, cp1, c="hotpink")
plt.title("Corriente vs Distancia Celda solar")
plt.xlabel("Distancia cm")
plt.ylabel("Corriente A")
plt.show()
```

```python
#Datos electrlisis

#Parte 1
vep1= np.array([1.3, 1.4, 1.5, 1.6, 1.7, 1.8])
iep1= np.array([0, 0, 0.02, 1.14, 0.24, 0.31])
```

```python
plt.scatter(vep1, iep1, c="forestgreen")
plt.title("Voltaje vs Corriente Electrolizador")
plt.xlabel("Voltaje V")
plt.ylabel("Corriente A")
plt.show()
```

```python
potencia1= (vp2[0])**2*ip2[0]
potencia2= (vp2[1])**2*ip2[1]
potencia3= (vp2[2])**2*ip2[2]
potencia4= (vp2[3])**2*ip2[3]
potencia5= (vp2[4])**2*ip2[4]
potencia6= (vp2[5])**2*ip2[5]
potencia7= (vp2[6])**2*ip2[6]
potencia8= (vp2[7])**2*ip2[7]
potencia9= (vp2[8])**2*ip2[8]

p= np.array([potencia1, potencia2, potencia3, potencia4, potencia5, potencia6, potencia7, potencia8, potencia9])

```

```python
plt.scatter(vp2, p, c="rebeccapurple")
plt.title("Potencia vs Voltaje Celda solar")
plt.ylabel("Potencia W")
plt.xlabel("Voltaje V")
plt.show()
```
