```python
import numpy as np
!pip install pint
!pip install pint uncertainties
import uncertainties
from uncertainties import ufloat
import pint as pint
ureg=pint.UnitRegistry(auto_reduce_dimensions=True)
Measurement=ureg.Measurement
Quantity=ureg.Quantity
```

```python
f = Quantity([1, 1.5, 2, 2.5198421], 'kHz')
f

L1 = Quantity([0, 17.5, 34, 52.5, 70.4], 'cm')

L1_1 = ufloat(L1[0].m, 0.5) * ureg.cm
L1_2 = ufloat(L1[1].m, 0.5) * ureg.cm
L1_3 = ufloat(L1[2].m, 0.5) * ureg.cm
L1_4 = ufloat(L1[3].m, 0.5) * ureg.cm
L1_5 = ufloat(L1[4].m, 0.5) * ureg.cm

L15 = Quantity([9.3, 20.5, 31.8, 47.7, 54.7], 'cm')

L15_1 = ufloat(L15[0].m, 0.5) * ureg.cm
L15_2 = ufloat(L15[1].m, 0.5) * ureg.cm
L15_3 = ufloat(L15[2].m, 0.5) * ureg.cm
L15_4 = ufloat(L15[3].m, 0.5) * ureg.cm
L15_5 = ufloat(L15[4].m, 0.5) * ureg.cm

L2 = Quantity([5.4, 13.8, 22.4, 31.2, 40], 'cm')

L2_1 = ufloat(L2[0].m, 0.5) * ureg.cm
L2_2 = ufloat(L2[1].m, 0.5) * ureg.cm
L2_3 = ufloat(L2[2].m, 0.5) * ureg.cm
L2_4 = ufloat(L2[3].m, 0.5) * ureg.cm
L2_5 = ufloat(L2[4].m, 0.5) * ureg.cm

L25 = Quantity([3.4, 9.8, 16.7, 23.6, 30.4], 'cm')

L25_1 = ufloat(L25[0].m, 0.5) * ureg.cm
L25_2 = ufloat(L25[1].m, 0.5) * ureg.cm
L25_3 = ufloat(L25[2].m, 0.5) * ureg.cm
L25_4 = ufloat(L25[3].m, 0.5) * ureg.cm
L25_5 = ufloat(L25[4].m, 0.5) * ureg.cm
```

```python
#Punto 1

n = np.array([1, 2, 3, 4, 5])

k1 = 2*n[0] - 1
k2= 2*n[1]-1
k3= 2*n[2]-1
k4= 2*n[3]-1
k5= 2*n[4]-1

print(k1, k2, k3, k4, k5)
```

```python
#Punto 2

t11= (k1/(4*f[0])).to('s')
t21= (k2/(4*f[0])).to('s')
t31= (k3/(4*f[0])).to('s')
t41= (k4/(4*f[0])).to('s')
t51= (k5/(4*f[0])).to('s')

t1= np.array([t11.magnitude, t21.magnitude, t31.magnitude, t41.magnitude, t51.magnitude])

t21= (k1/(4*f[1])).to('s')
t22= (k2/(4*f[1])).to('s')
t23= (k3/(4*f[1])).to('s')
t24= (k4/(4*f[1])).to('s')
t25= (k5/(4*f[1])).to('s')

t2= np.array([t21.magnitude, t22.magnitude, t23.magnitude, t24.magnitude, t25.magnitude])

t31= (k1/(4*f[2])).to('s')
t32= (k2/(4*f[2])).to('s')
t33= (k3/(4*f[2])).to('s')
t34= (k4/(4*f[2])).to('s')
t35= (k5/(4*f[2])).to('s')

t3= np.array([t31.magnitude, t32.magnitude, t33.magnitude, t34.magnitude, t35.magnitude])

t41= (k1/(4*f[3])).to('s')
t42= (k2/(4*f[3])).to('s')
t43= (k3/(4*f[3])).to('s')
t44= (k4/(4*f[3])).to('s')
t45= (k5/(4*f[3])).to('s')

t4= np.array([t41.magnitude, t42.magnitude, t43.magnitude, t44.magnitude, t45.magnitude])
```

```python
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit

def ajuste (x, a, b):
  return a*x + b

L1_m= L1.magnitude

poct, pocv= curve_fit(ajuste, t1, L1_m)

plt.plot(t1, L1_m)
plt.plot(t1, ajuste(t1,*poct), '--')
plt.title('V1')
plt.xlabel('t1')
plt.ylabel('L1')
plt.show()

err = np.sqrt(np.diag(pocv))

print(f'Velocidad 1=', poct[0])
```

```python
from matplotlib import pyplot as plt

def ajuste (x, a, b):
  return a*x + b

L15_m= L15.magnitude

poct2, pocv2= curve_fit(ajuste, t2, L15_m)

plt.plot(t2, L15_m)
plt.plot(t2, ajuste(t2, *poct2), '--')
plt.title('V2')
plt.xlabel('t2')
plt.ylabel('L15')
plt.show()

err2 = np.sqrt(np.diag(pocv2))

print(f'Velocidad 2=', poct2[0])
```

```python
from matplotlib import pyplot as plt

def ajuste (x, a, b):
  return a*x +b

L2_m= L2.magnitude

poct3, pocv3= curve_fit(ajuste, t3, L2_m)


plt.plot(t3, L2, c='k')
plt.plot(t3, ajuste(t3, *poct3), '--', c= 'orange')
plt.title('V3')
plt.xlabel('t3')
plt.ylabel('L2')
plt.show()

err3 = np.sqrt(np.diag(pocv3))

print(f'Velocidad 3=', poct3[0])
```

```python
#Aquí pls

from matplotlib import pyplot as plt

def ajuste (x, a, b):
  return a*x +b

L25_m= L25.magnitude

poct4, pocv4= curve_fit(ajuste, t4, L25_m)

plt.plot(t4, L25_m)
plt.plot(t4, ajuste(t4, *poct4), '--')
plt.title('V4')
plt.xlabel('t4')
plt.ylabel('L25')
plt.show()

#incertidumbre curve_fit

err4 = np.sqrt(np.diag(pocv4))


print(f'Velocidad 4=', poct4[0], f'+-', err4[0])
```
