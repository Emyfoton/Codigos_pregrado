```python
#Datos primero
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definir los datos
data = {
    "Thielo": [16.0, 16.0, 16.5, 17.0, 18.0, 18.0, 18.5, 20.0, 20.5, 21.5, 22.0, 24.0],
    "Tcaliente": [38, 39.5, 41, 42, 44, 45, 46, 48, 49, 51, 52, 54.5],
    "Voltaje (Volts)": [0.9, 0.95, 0.98, 1.01, 1.04, 1.07, 1.1, 1.14, 1.15, 1.19, 1.21, 1.25],
    "Corriente (mA)": [75.1, 79, 82.1, 86, 91.5, 95.1, 97.9, 101.5, 109.4, 112.7, 115.5, 118.5]
}
# Crear el DataFrame
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

```python
Th = np.array(df['Thielo'])
Tc = np.array(df['Tcaliente'])
voltajes = np.array(df['Voltaje (Volts)'])
corriente = np.array(df['Corriente (mA)'])

dif = Tc - Th
```

```python
plt.plot(dif,voltajes, c='black')
plt.title('Corriente vs Temperatura')
plt.xlabel('Î” Temperatura (K)')
plt.ylabel('Voltaje (V)')
plt.show()
```

```python
from scipy.optimize import curve_fit

def ajuste (x, a, b):
  return a*x +b

popt, pocv= curve_fit(ajuste, dif, voltajes)

plt.plot(dif,voltajes, c= 'k')
plt.plot(dif, ajuste(dif, *popt), '--', c='red')
plt.xlabel('Î” Temperatura (CÂ°)')
plt.ylabel('Voltaje (V)')
plt.title('Ajuste Corriente vs Temperatura')
plt.legend(['Datos', 'Ajuste'])
plt.show()

err = np.sqrt(np.diag(pocv))

print(f'Ajuste', popt)
print(f'error', err)
```

```python
seebeck = ufloat(popt[0], err[0]) * ureg.volt / ureg.K
seebeck
```
