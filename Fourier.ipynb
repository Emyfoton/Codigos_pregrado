```python
import pandas as pd

ruta = "Ondas normalitas (S-C-T).xlsx"

filas_deseadas = [3, 4]

df = pd.read_excel(ruta)
df_filtrado = df.iloc[:, filas_deseadas]

print(df_filtrado)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= df.iloc[:, 3]
y= df.iloc[:, 4]

plt.plot (x,y, c= "olive")
plt.title("Onda Sinusoidal")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()

```

```python
# la FFT
from scipy.signal import find_peaks

tiempo = np.array(df.iloc[:, 3], dtype=float)
carga = np.array(df.iloc[:, 4], dtype=float)


N = len(carga)  # Número de puntos
fft_carga = np.fft.fft(carga)  # Transformada de Fourier
frecuencias = np.fft.fftfreq(N, d=(tiempo[1] - tiempo[0]))  # Espaciado temporal


# Graficar el espectro de Fourier y marcar los picos
plt.plot(frecuencias[:N//2], np.abs(fft_carga[:N//2]), c= "olive")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.title("Fourier de la Señal sinusoidal")
plt.grid(True)
plt.xlim(0, 200)

plt.show()
```

```python
#Gráfica función numérica de suma con ruído

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000, endpoint=False)

funcion= 10*np.cos(2*np.pi*10*t)
plt.plot(t,funcion)
plt.title("Gráfica Acos(2πωt)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```
